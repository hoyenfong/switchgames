```python
top_5=switchdf.head(5)
```

```python
df = pd.DataFrame(top_5)
```

```python
df.sort_values(by='copies_sold', ascending=True, inplace=True)
```

```python
plt.barh(df['title'], df['copies_sold'], color='orange')
plt.xlabel('Copies Sold')
plt.ylabel('Game')
plt.title('Top 5 Best Selling Nintendo Switch Games')
plt.gca().get_xaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.show()
```
